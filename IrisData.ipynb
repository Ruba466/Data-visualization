```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(iris["sepal_width"],bins=20,kde=True,color="lightpink",edgecolor="black")
plt.title("histogram sepal width")
plt.xlabel("sepal width")
plt.ylabel("frequency")

plt.subplot(1,2,2)
sns.boxplot(x="sepal_width",data=iris,color="lightgreen")
plt.title("boxplot sepal width")
plt.xlabel("species")
plt.ylabel("sepal_width")
plt.tight_layout()
plt.show()  
```
