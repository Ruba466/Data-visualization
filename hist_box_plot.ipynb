```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv("tips.csv")
tips.head()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Tips Dataset Visualizations', fontsize=16)

#histogram
sns.histplot(tips['total_bill'], bins=20, color='lightgreen', edgecolor='black', ax=axs[0, 0])
axs[0, 0].set_title("Histogram of Total Bill")
axs[0, 0].set_xlabel("Total Bill")
axs[0, 0].set_ylabel("Frequency")

#boxplot
sns.boxplot(y=tips["tip"], color='skyblue', ax=axs[0, 1])
axs[0, 1].set_title("Boxplot of Tip amount")
axs[0, 1].set_ylabel("Tip ($)")

#piechart
gender_counts = tips['sex'].value_counts()
axs[1, 0].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightpink', 'lightblue'],wedgeprops={'edgecolor': 'black'})
axs[1, 0].set_title("Gender Distribution")
axs[1, 0].axis('equal')

#barchart
avg_total_by_day = tips.groupby('day')['total_bill'].mean().sort_index()
sns.barplot(x=avg_total_by_day.index,y=avg_total_by_day.values, color="orange", ax=axs[1, 1])
axs[1, 1].set_title("Avg Total Bill by Day")
axs[1, 1].set_xlabel("Day")
axs[1, 1].set_ylabel("Avg Total Bill")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
